```python
df['Author(s) ID'] = df['Author(s) ID'].str.rstrip(';')
```

```python
df_authors = df_authors.assign(Authors=df_authors['Authors'].str.split(',')).explode('Authors').reset_index(drop=True)
df_authors.drop(132, inplace=True)
df_authors.reset_index(drop=True, inplace=True)
df_authors.head(10)
```

#### The row with modified-index 132 was dropped because the name was mistankingly added to the dataset and didn't belong to anybody.

# Getting the trend in Source Title

```python
df_st = df.copy()
df_st["year_groups"] = pd.cut(df_st["Year"], 
                             bins = [1981, 2006, 2011, 2015, np.inf], 
                             labels = ["1981-2005", "2006-2010", "2011-2014", "2015-2019"])


df_st["Source title"] = df["Source title"].astype(str)
df_trends = df_st[['year_groups', 'Source title']].groupby("year_groups").sum()

#Resetting the index
df_st = df_st.reset_index()
df_st["year_groups"].value_counts()
```

Distribution of the character used in Source title

```python
df_st["Source title"].str.len().hist()
plt.xlabel("Number of characters per Source title")
plt.ylabel("Frequency")
```

Number of words used per Source title

```python
df_st["Source title"].str.split().map(lambda x: len(x)).hist() #This line converts the text source title into a list of slit words and counts the words with the len function
plt.xlabel("Number of Words per Source title")
plt.ylabel("Frequency")
```

```python
df_trends = df_trends.reset_index()
df_trends
```
